Write a Python program to remove and print every third number from a list of numbers until the list becomes empty


Liste boşalana kadar bir sayı listesinden her üçüncü sayıyı çıkart ve bunu yazdıran bir Python programı yazın

example :
image.png

```python
liste = [1 ,2 ,3 ,4,5 ,6 ,7 ,8 ,9]
liste2=[]
index = 0 
length = len(liste)

while len(liste) > 0 :
  index = (2 + index) % length
  x =liste.pop(index)
  print(liste) 
  liste2.append(x)
  length -= 1
print(liste2)  
```

```python
liste = [1 ,2 ,3 ,4 ,5 ,6 ,7 ,8 ,9]
index = 0 
length = len(liste)

while len(liste) > 0 :
  index = (2 + index) % length
  liste.pop(index)
  print(liste)  
  length -= 1
```
