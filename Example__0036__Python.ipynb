Write a Python function that takes a sequence of numbers and determines whether all the numbers are different from each other


Bir sayı dizisi alan ve tüm sayıların birbirinden farklı olup olmadığını belirleyen bir Python işlevi yazın.

image.png

```python
number = input("Enter a number = ")

print(len(set(number)) == len(number))    # set tekrar eden ikinci bir şeyi almaz.
```
