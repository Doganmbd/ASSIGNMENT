```python
# Print the times table from 1 to 10 as in the example

j = 1
while j <= 10 :
  for i in range(1,11) :
    if len(str(i))==1 and len(str(j))==1 and len(str(i*j))==1:
      print(" {} x  {} =   {}".format(i,j,i*j), end = "|")
    elif len(str(i))==1 and len(str(j))==1 and len(str(i*j))==2:
      print(" {} x  {} =  {}".format(i,j,i*j), end = "|")
    elif len(str(i))==2 and len(str(j))==1 and len(str(i*j))==2:
      print("{} x  {} =  {}".format(i,j,i*j), end = "|")
    elif len(str(i))==1 and len(str(j))==2 and len(str(i*j))==2:
      print(" {} x {} =  {}".format(i,j,i*j), end = "|")
    elif len(str(i))==2 and len(str(j))==2 and len(str(i*j))==3:
      print("{} x {} = {}".format(i,j,i*j), end = "|")
    else:
      print(" {} x {} ={}".format(i,j,i*j), end = "|")
  j += 1
  print()
```

```python
# If you do it this way, the shape will be distorted. And as in the example, the numbers will appear side by side, not one under the other.

for i in range(1,11):
  for j in range(1,11):
    print("{} X {}  =  {}    ".format(i,j,j * i),end="|")
  print()  
  
```
