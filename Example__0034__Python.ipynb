Write a Python program to remove the characters which have odd index values of a given string


Belirli bir dizgenin tek indeks değerlerine sahip karakterleri kaldırmak için bir Python programı yazın

For Example :

image.png

```python
text = input("Enter a text = ")
word = ""

for i in range(len(text)):
  if i % 2 == 0:
    word += text[i]
print(word)  
```
